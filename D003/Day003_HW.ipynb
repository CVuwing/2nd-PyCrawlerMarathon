### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


File I/O讀出來應該就是該檔案的原始資料，以字串的方式顯示，xmltodict會讀取的xml轉換成dictionary的型態，而dictionary中又會由Tuple () List[]和Dictionary {}組成內容，方便找尋資料

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 讀檔案
import xmltodict
fh = open("./data/64_72hr_CH.xml","r",encoding="utf-8")
xml = fh.read()
doc = dict(xmltodict.parse(xml))
print(doc)
```

```python
# 1. 請問高雄市有多少地區有溫度資料？
i=0;
areas = doc['cwbopendata']['dataset']['locations']['location']
# print(areas)
for area in areas:
     print (area['locationName'])
     i=i+1
print(i,'個')
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
for area in areas:
    print (area['weatherElement'][0]['time'][0]['dataTime'],area['weatherElement'][0]['time'][0]['elementValue']['value'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
times = areas[0]['weatherElement'][0]['time']
for time in times:
    print (time['dataTime'],time['elementValue']['value'])
```

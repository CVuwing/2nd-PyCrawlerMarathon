### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")
soup =soup.find("div", class_= "r-ent")
print(soup)
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

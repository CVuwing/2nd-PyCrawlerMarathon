## 1. 取出今天所有的新聞

```python
import time
import datetime
# now = int(time.time()) 
# # print(time.time())
# # print(time.time()- datetime.timedelta(days=3))
# threeDayAgo = datetime.timedelta(days=-3)
# print(threeDayAgo)
# print(datetime.datetime.now().date())
# print(datetime.datetime.now().date() - datetime.timedelta(days=3))
current_time=datetime.datetime.now()
current_timestamp = datetime.datetime.timestamp(current_time)
```

```python
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

# 每個兩秒鐘自動往下滑

import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();
```

## 2. 取出現在時間兩小時內的新聞

```python

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

# 每個兩秒鐘自動往下滑

import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
in2hour_time_stamp = int(datetime.datetime.timestamp(current_time-datetime.timedelta(hours=2)))
for d in soup.find(class_="part_list_2").find_all('h3'):
    time_str = d.find(class_="date").text
    time_stamp = int(time.mktime(datetime.datetime.strptime(time_str, "%Y/%m/%d %H:%M").timetuple()))
    if(time_stamp >= in2hour_time_stamp):
        print(time_str, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
current_date_str=(current_time - datetime.timedelta(days=3)).strftime("%Y-%m-%d")
browser.get("https://www.ettoday.net/news/news-list-"+current_date_str+"-0.htm")
fromTime_str=current_date_str+" 15:00"
toTime_str=current_date_str+" 17:00"
fromTimestamp = int(time.mktime(datetime.datetime.strptime(fromTime_str, "%Y-%m-%d %H:%M").timetuple()))
toTimestamp = int(time.mktime(datetime.datetime.strptime(toTime_str, "%Y-%m-%d %H:%M").timetuple()))
# 每個兩秒鐘自動往下滑

import time
for i in range(50):
    time.sleep(0.5)
    browser.execute_script("window.scrollTo(0, 50000);")

# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    time_str = d.find(class_="date").text
    time_stamp = int(time.mktime(datetime.datetime.strptime(time_str, "%Y/%m/%d %H:%M").timetuple()))
    if(time_stamp >= fromTimestamp and time_stamp <= toTimestamp):
        print(time_str, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();
```
